```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import math
```

```python
def historico(ticker):
    activo = yf.Ticker(ticker)
    return activo.history(period = "max")
```

```python
dataframes = {}
```

```python
tickers_tec = ['AAPL', 'MSFT', 'GOOGL', 'NVDA', 'META', 'ADBE', 'ORCL', 'CRM', 'INTC', 'AMD']
tickers_com = ['AMZN', 'WMT', 'COST', 'EBAY', 'TGT', 'JD', 'BABA', 'HD', 'LOW', 'M']
tickers_fin = ['JPM', 'BAC', 'GS', 'WFC', 'C', 'AXP', 'V', 'MA', 'BRK-B', 'MS']
tickers_ind = ['BA', 'CAT', 'MMM', 'GE', 'HON', 'DE', 'LMT', 'UPS', 'FDX', 'RTX']
todos = tickers_tec + tickers_com + tickers_fin + tickers_ind
for activo in tickers_tec + tickers_com + tickers_fin + tickers_ind:
    dataframes[activo] = historico(activo)
```

```python
def retornos(df):
    df['Retornos'] = np.log(df['Close']/df['Close'].shift(1))
    return df

for activo in dataframes.keys():
    dataframes[activo] = retornos(dataframes[activo])
```

```python
import pandas as pd

retornos = {}

# Itera sobre cada acción en el diccionario
for accion in dataframes.keys():
    retornos[accion]=dataframes[accion]["Retornos"]


retornos =pd.DataFrame(retornos)
retornos
```

```python
# Calcula la matriz de correlación
matriz_correlacion = retornos.corr()
sns.heatmap(matriz_correlacion)
```

```python
def calculo_parametros(df, name = str):
    media = df['Retornos'].mean()
    volatilidad = df['Retornos'].std()
    data = {
        f'Media {name}': [media, media*20, media*250],
        f'Volatilidad {name}':[volatilidad, volatilidad*(20**(1/2)), volatilidad*(250**(1/2))]
        }
    return pd.DataFrame(data, index=['Diario', 'Mensual', 'Anual']).transpose()


parametros = pd.DataFrame()

for activo in dataframes:
    df = calculo_parametros(dataframes[activo], activo)
    parametros = pd.concat([parametros, df], axis = 0)
```

```python
def precio_futuro(precio_spot, promedio_diario, volatilidad_diaria, normalidad):
    return precio_spot*math.e**(promedio_diario*1+ volatilidad_diaria*math.sqrt(1)*normalidad)

def pronostico(ticker, cantidad_dias, cantidad_pronosticos):
    array = np.array([[float(dataframes[ticker]['Close'][-1]) for i in range(cantidad_pronosticos)]])

    for i in range(cantidad_dias):
        
        resultado = np.array([precio_futuro(float(spot), #precio actual
                                float(parametros.loc[f'Media {ticker}', 'Diario']), #media
                                float(parametros.loc[f'Volatilidad {ticker}', 'Diario']), #volatilidad
                                float(np.random.randn(1)[0]))#numero aleatorio de la distribución normal 
                                for spot in array[-1]])#selecciona el ultimo resultado y calcula uno nuevo

        array = np.append(array, [resultado], axis = 0)
    return array
```

```python
pronosticos = {}
for activo in dataframes.keys():
    pronosticos[activo] = pd.DataFrame(pronostico(activo, 90, 2000))
```

```python
def max_min_PnL(ticker, diccionario):
    media = diccionario[ticker].transpose().mean()
    volatilidad = diccionario[ticker].transpose().std()
    ls = media+(volatilidad*2)
    li = media-(volatilidad*2)
    return pd.DataFrame({'Media': media,
            'lim sup': ls,
            'lim inf': li})

max_min = {}
for activo in pronosticos:
    max_min[activo] = max_min_PnL(activo, pronosticos)
```

```python
fig, axes = plt.subplots(8, 5, figsize=(15, 20))
a=0
for i, ax in enumerate(axes.flatten()):
    
    sns.lineplot(data=max_min[todos[a]], ax=ax)
    ax.set_title(f'Gráfico {todos[a]}')  
    a+=1
```

```python
def porcentajes(activo):
    min_porcentaje = (1 - max_min[activo]['lim inf'][90]/max_min[activo]['Media'][0])*100
    max_porcentaje = (max_min[activo]['lim sup'][90]/max_min[activo]['Media'][0] - 1)*100
    return min_porcentaje, max_porcentaje

porcentaje_perdida_ganancia = pd.DataFrame({}, index = ['% perdida máxima', '% ganacia máxima'])

for activo in max_min.keys():
    min_porcentaje, max_porcentaje = porcentajes(activo)
    porcentaje_perdida_ganancia[activo] = [min_porcentaje, max_porcentaje]

porcentaje_perdida_ganancia = porcentaje_perdida_ganancia.transpose()

porcentaje_perdida_ganancia['diferencia gan-per'] = porcentaje_perdida_ganancia['% ganacia máxima']-porcentaje_perdida_ganancia['% perdida máxima']
```
